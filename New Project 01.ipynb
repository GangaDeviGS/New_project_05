First, we can check the relationship between 'year' and 'battle_number'.

```python
dataA = data['year']
dataB = data['battle_number']

plt.bar(dataA,dataB,color='red',width=0.1)
plt.xlabel('Year')
plt.ylabel('battle_number')
plt.title('Bar Plot')
plt.show()
```

This bar plot is very easy to understand. For a quick look, we can see that most battles occur in 300 AC. This is the benefit of visualization.

```python
plt.figure(figsize=(10,4))
sns.countplot(x=data['attacker_king'],hue=data['battle_type'] , width =0.4)
plt.xlabel('Attacker King')
plt.title('Battle Type')
plt.show()
```

In this countplot, we can see what the battle type is followed by the attacker king. Without any doubt, we can say that Joffrey/Tommen Baratheon's battle type was pitched battle. Robb Stark's battle type was ambush. Balon/Euron Greyjoy was the only one King  done Razing battle.

```python
plt.figure(figsize=(15,4))
data.plot(kind='bar',x='attacker_1',y='attacker_size',color = 'green',width = 0.5)
plt.show()
```
