**KNN**

K Nearest Neighbors is one of the simplest Supervised Machine Learning algorithm mostly used for classification. It classifies a data point based on how its neighbors are classified.

```python
# Import the required libraries

import numpy as np
import pandas as pd
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
# Import Dataset

data = pd.read_csv('/content/diabetes.csv')
```

Values of columns like 'Glucose', 'BloodPressure' cannot be accepted as zeroes because it will affect the outcome. We can replace such values with the mean of the respective column

```python
# Replace zeroes

replace_zeroes = ['Glucose', 'BloodPressure', 'SkinThickness', 'BMI', 'Insulin']
```

```python
for column in replace_zeroes :
  data[column] = data[column].replace(0, np.NaN)
  mean = int(data[column].mean(skipna = True))
  data[column] = data[column].replace(np.NaN, mean)
```

```python
data.describe()
```

All the 0's have been replaced by mean

```python
# Split the dataset into train and test

X = data.iloc[:, 0:8]
Y = data.iloc[:, 8]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state = 0, test_size = 0.2)
```

**Feature Scaling**

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

Define the model using KNeighborsClassifier and fit the train data in the model.N_neighbors here is 'K' p is the power parameter to define the metric used. Which is 'Euclidean' in our case

```python
import math
math.sqrt(len(Y_test))
```

In here the square root of the sum of the squared differences between the coordinates of two points is a even number. To avoid ties, it is recommended to use an odd value of k

```python
# Define the model : Init KNN

classifier = KNeighborsClassifier(n_neighbors = 11, p = 2, metric = 'euclidean')
```

It's important to evaluate the model, Let's use confusion matrix to do that

```python
# Fit Model
classifier.fit(X_train, Y_train)
```

```python
Y_predict = classifier.predict(X_test)
Y_predict
```

```python
# Evaluate Model

cm = confusion_matrix(Y_test, Y_predict)
print (cm)
```

```python
print (f1_score(Y_test, Y_predict))
```

```python
print (accuracy_score(Y_test, Y_predict))
```

We have developed a KNN model that can predict if a person will develop diabetes or not, or at the very least, if they should get checked out and have their blood sugar checked on a regular basis. The accurate score we obtained, 0.80519, was fairly close to what we obtained, so we can pretty much round that off and say that our accuracy of 80% indicates that the model fits the data fairly well.
