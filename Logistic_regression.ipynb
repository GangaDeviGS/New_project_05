```python
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv("/content/Heart (1).csv")
df
```

```python
df.shape
```

```python
df.head()
```

```python
df.tail()
```

```python
# Remove unwanted column
df.drop(columns = 'Unnamed: 0')
```

```python
# Convert categorical column into numerical codes
df['ChestPain'] = df['ChestPain'].astype('category')
df['ChestPain'] = df['ChestPain'].cat.codes
df
```

```python
df['Thal'] = df['Thal'].astype('category')
df['Thal'] = df['Thal'].cat.codes

df['AHD'] = df['AHD'].astype('category')
df['AHD'] = df['AHD'].cat.codes

df
```

```python
# Check null values
df.isnull().any()
```

```python
# count of null values
df.isnull().sum()
```

There are four null values. So that we can get rid of it

```python
df = df.dropna()
df
```

```python
df.isnull().sum()
```

```python
# Divide the dataset into two parts

X = df.drop(columns = 'AHD')
X
```

```python
Y = df['AHD']
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_test, X_train, Y_test, Y_train = train_test_split(X, Y, test_size = 0.3, random_state = 25)
```

```python
X_train
```

```python
X_test, Y_train, Y_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaler = scaler.fit_transform(X_train)
X_train_scaler
```

```python
X_test_scaler = scaler.fit_transform(X_test)
X_test_scaler
```

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression(random_state = 0).fit(X_train_scaler, Y_train)
```

```python
lg.predict(X_train_scaler)
```

```python
lg.predict(X_test_scaler)
```

```python
lg.score(X_train_scaler, Y_train)
```

```python
lg.score(X_test_scaler, Y_test)
```
